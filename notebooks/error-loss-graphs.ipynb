## Plots graphs of error loss values for different metrics

#### Plotted values are for a singular label/prediciton, but for a large set of predictions the error loss is simply the mean of each individual data point's error loss

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
prediction = np.arange(0.001,1,0.001)
brier = (1-prediction)**2
logloss = -np.log(prediction)
```

```python
plt.figure()
plt.figure(figsize=(10, 10))
plt.rc('font', size=10)          # controls default text sizes
plt.rc('axes', titlesize=30)     # fontsize of the axes title
plt.rc('axes', labelsize=30)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=30)    # fontsize of the tick labels
plt.rc('ytick', labelsize=30)    # fontsize of the tick labels
plt.rc('legend', fontsize=30)    # legend fontsize
plt.rc('figure', titlesize=30)  # fontsize of the figure title
plt.xlabel('Predicted Probability')
plt.ylabel('Score')
plt.plot(prediction, brier, label = 'brier')
plt.plot(prediction, logloss, label = 'cross-entropy loss')
plt.legend()
plt.ylim(-0.2,1.8)
plt.tight_layout()
plt.savefig('graphs/error-loss.png')
plt.show()
```
